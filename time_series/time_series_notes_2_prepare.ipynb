```python
df = (sales
            .merge(
                items, 
                how='outer', 
                left_on='item', 
                right_on='item_id', 
                validate='m:1'
            )
            .merge(
                stores, 
                how='outer', 
                left_on='store', 
                right_on='store_id', 
                validate='m:1'
            )
           )
frame_splain(df, splain=True)
```
